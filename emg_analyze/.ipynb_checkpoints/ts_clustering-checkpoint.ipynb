## ライブラリのインポート

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os

from tslearn.utils import to_time_series
from tslearn.clustering import TimeSeriesKMeans
from tslearn.clustering import KShape
from tslearn.preprocessing import TimeSeriesScalerMeanVariance

%matplotlib inline
```

tslearnパッケージについては、下記URLを参照。

インストール方法　：　https://tslearn.readthedocs.io/en/latest/index.html

## ファイルの読み込み

```python
#読み込むデータ一覧を取得する。
#データは他のデータセットから筋電位として抽出できたと仮定したデータをcsvにしてフォルダーにまとめておきました。
files = glob.glob("../emg_analyze/data/*.csv")

df_list = []

for path in files:
    df = pd.read_csv(path)
    df_list.append(df)

#データを連結する
ftd = pd.concat(df_list).drop("Unnamed: 0", axis=1)
ftd.info()
ftd.head()
```


ここで読み込んだデータは各実験のipynbファイルで、周波数帯域がgamma2以上のもの、EEG_analyzeツールで使われていた表記で書くと、

df_fft[emg] = cls_dataset.sum_hz(50.39,59.77)

として筋電位として抽出し、結合したものである。。（上記のコードを各実験のipynbファイルで行っていただいて、to_csv()で抽出したものを結合している。）

今回はまだ完璧に脳波と筋電が分離できたと断言はできないが、これで仮に分離できたと仮定して分析を進める。

# 前処理

```python
#実験リストの取得
ExperNo_list = ftd.管理No.unique()
print(ExperNo_list)
```

・実験番号の概要

FTD120010 ~ FTD120014 : つばを飲み込むテスト

FTD130010 ~ FTD130014 : 口を開けるテスト

FTD131010 ~ FTD131014 : 噛むテスト

FTD150010 ~ FTD150014 : ほおをあげる

## 学習データセットの時系列の長さを揃える。（今回は簡略化のため、一番短い長さのものに合わせる。）

```python
#今回は筋電位のデータだけ使用する。
ftd_emg = ftd[["管理No", "emg"]]

ftd_emg_len = []
ftd_emg_list = []

#各実験の実験時間をリストとして取得。
for ExperNo in ExperNo_list:
    ftd_emg_len.append(len(ftd_emg.emg[ftd_emg.管理No==ExperNo][:]))

#その中で一番短い実験時間を取得。
min_length = min(ftd_emg_len)

#核実験について、同じ長さに統一した。
for ExperNo in ExperNo_list:
    ftd_emg_list.append(ftd_emg.emg[ftd_emg.管理No==ExperNo][:min_length])

#tslearnデータセットのフォーマットに合わせた。
formatted_dataset = to_time_series(ftd_emg_list)
print("時系列データセット数：{0} \n時系列の長さ(秒):{1}".format(formatted_dataset.shape[0], formatted_dataset.shape[1]))
```

# 分析

それぞれの分析手法についてのドキュメントは下記を参考にしてください。

k-means : https://tslearn.readthedocs.io/en/latest/gen_modules/clustering/tslearn.clustering.TimeSeriesKMeans.html

k-shape : https://tslearn.readthedocs.io/en/latest/gen_modules/clustering/tslearn.clustering.KShape.html

## k-meansを用いたクラスタリング

今回はtslearnのk-meansに実装されている中で距離の測定基準として、

・DTW

・soft-DTW

・euclidean

を全て試してみた。
これらの測定基準の違いについてはまだ理解していないので、後日記載します。

```python
#dtw
km = TimeSeriesKMeans(n_clusters=4, verbose=False ,metric="dtw")
labels = km.fit_predict(formatted_dataset)
print("result:{0}".format(labels))

plt.figure(figsize=(40,20))
for yi in range(4):
    plt.subplot(3, 4, yi + 1)
    for xx in formatted_dataset[labels == yi]:
        plt.plot(xx.ravel(), "k-", alpha=.2)
    plt.plot(km.cluster_centers_[yi].ravel(), "r-")
    plt.ylim(-10, 30)
    if yi == 1:
        plt.title("dtw")

#softdtw(softdtwについては時系列の長さが違うものをデータセットに入れてみても実行できる。)
km = TimeSeriesKMeans(n_clusters=4, metric="softdtw", verbose=False)
labels = km.fit_predict(formatted_dataset)

print("result:{0}".format(labels))
plt.figure(figsize=(40,20))
for yi in range(4):
    plt.subplot(3, 4, yi + 1)
    for xx in formatted_dataset[labels == yi]:
        plt.plot(xx.ravel(), "k-", alpha=.2)
    plt.plot(km.cluster_centers_[yi].ravel(), "r-")
    plt.ylim(-10, 30)
    if yi == 1:
        plt.title("softdtw")
        
##ユークリッド
km = TimeSeriesKMeans(n_clusters=4, metric="euclidean", verbose=False)
labels = km.fit_predict(formatted_dataset)

print("result:{0}".format(labels))
plt.figure(figsize=(40,20))
for yi in range(4):
    plt.subplot(3, 4, yi + 1)
    for xx in formatted_dataset[labels == yi]:
        plt.plot(xx.ravel(), "k-", alpha=.2)
    plt.plot(km.cluster_centers_[yi].ravel(), "r-")
    plt.ylim(-10, 30)
    if yi == 1:
        plt.title("euclidean")
```

## k-shapeを用いたクラスタリング

```python
#実験結果が変わると考察しにくいので、今回はrandom_stateを0に固定しておいた。
ks = KShape(n_clusters=4, verbose=False, random_state=0)
y_pred = ks.fit_predict(formatted_dataset)

print("result:{0}".format(labels))
plt.figure(figsize=(20,10))
for yi in range(4):
    plt.subplot(4, 1, 1 + yi)
    for xx in formatted_dataset[y_pred == yi]:
        plt.plot(xx.ravel(), "k-", alpha=.2)
    plt.plot(ks.cluster_centers_[yi].ravel(), "r-")
    plt.title("Cluster %d" % (yi + 1))

plt.tight_layout()
plt.show()
```
